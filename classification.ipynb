```python
import pandas as pd
diabetes = pd.read_csv('diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
diabetes
```

```python
diabetes.columns
```

```python
pd.set_option('display.max_columns', None)
diabetes
```

```python
feature_cols = ['HighBP', 'HighChol', 'CholCheck', 'BMI', 'Smoker',
       'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
       'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth',
       'MentHlth', 'PhysHlth', 'DiffWalk', 'Sex', 'Age', 'Education',
       'Income']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

from scipy.stats import sem

feature_cols = [ col for col in feature_cols if col != 'Diabetes_binary' ]
diabetes_tree = DecisionTreeClassifier(random_state=0, criterion='entropy', max_depth=3, min_samples_leaf=50, class_weight='balanced')
cv_score = cross_val_score(diabetes_tree, 
                           diabetes.loc[:, feature_cols], 
                           diabetes['Diabetes_binary'], cv=10,
                           scoring='accuracy')

print(np.mean(cv_score), sem(cv_score))
```

```python
diabetes_tree.fit(diabetes.loc[:, feature_cols], diabetes['Diabetes_binary'])
```

```python
print(f"Tree depth: {diabetes_tree.get_depth()}")
print(f"Number of leaves: {diabetes_tree.get_n_leaves()}")
```

```python
from sklearn import tree
import graphviz
from IPython import display  

dot_data = tree.export_graphviz(diabetes_tree, out_file=None,  
                                feature_names=feature_cols,  
                                class_names=["false", "true"],  
                                filled=True, rounded=True,  
                                special_characters=True) 

graph = graphviz.Source(dot_data)  
display.display_png(graph)
```

```python
from sklearn.ensemble import RandomForestClassifier

diabetes_rf = RandomForestClassifier(n_estimators=50, max_depth=None, min_samples_split=2, random_state=0)
cv_score = cross_val_score(diabetes_rf, 
                           diabetes.loc[:, feature_cols], 
                           diabetes['Diabetes_binary'], cv=10,
                           scoring='accuracy')
cv_score.mean()
```

```python
X = diabetes.loc[:, feature_cols]
y =  diabetes['Diabetes_binary']

diabetes_rf.fit(X, y)
importances = diabetes_rf.feature_importances_

std = np.std([tree.feature_importances_ for tree in diabetes_rf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

for f in range(X.shape[1]):
    print("%d. feature %d %s (%f)" % (f + 1, indices[f], feature_cols[indices[f]], importances[indices[f]]))
```

```python
import matplotlib.pyplot as plt

# Select top 10 features
top_n = 10
indices_top = indices[:top_n]  # Get the indices of the top 10 features
importances_top = importances[indices_top]  # Corresponding importances
std_top = std[indices_top]  # Corresponding standard deviations
features_top = [feature_cols[x] for x in indices_top]  # Names of top 10 features

# Plot the feature importances of the forest (top 10)
plt.figure(figsize=(9, 5))
plt.title("Top 10 Feature Importances")
plt.bar(range(top_n), importances_top,
        tick_label=features_top,
        color="r", yerr=std_top, align="center")

# Set x-axis limit and rotate labels for better visibility
plt.xlim([-1, top_n])
plt.xticks(rotation=45)
plt.show()
```

# CatBoost

```python
from catboost import CatBoostClassifier

Selected_features = ['GenHlth' ,'CholCheck' ,'HighBP' ,'AnyHealthcare' ,'PhysActivity' ,'BMI' ,
'HighChol' ,'Age' ,'Fruits' ,'Income' ,'DiffWalk' ,'Education' ,'HeartDiseaseorAttack'
,'PhysHlth' , 'Smoker']
Target = 'Diabetes_binary'

X = diabetes[Selected_features]
y = diabetes[Target]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size=0.2)
cat  = CatBoostClassifier(verbose=0)
```

```python
# Train with the resampled data
cat.fit(X_train, y_train)

y_pred = cat.predict(X_test)
```

```python
# Assuming y_pred contains predicted probabilities
y_pred_prob = cat.predict(X_test)  # Get predicted probabilities

# Convert probabilities to binary labels using 0.5 as the threshold
y_pred = [1 if prob > 0.5 else 0 for prob in y_pred_prob]

# Now evaluate using classification metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

# Precision
precision = precision_score(y_test, y_pred)
print(f'Precision: {precision}')

# Recall
recall = recall_score(y_test, y_pred)
print(f'Recall: {recall}')

# F1 Score
f1 = f1_score(y_test, y_pred)
print(f'F1 Score: {f1}')

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(f'Confusion Matrix:\n{conf_matrix}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'cat' is your trained CatBoost model
cat_feature_importances = cat.get_feature_importance()

# Assuming 'Selected_features' is your list of feature names
feature_names = np.array(Selected_features)

# Sort the feature importances in descending order
sorted_indices = np.argsort(cat_feature_importances)[::-1][:10]

# Plot the top 10 features
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(range(10), cat_feature_importances[sorted_indices], align='center')
ax.set_yticks(range(10))
ax.set_yticklabels(feature_names[sorted_indices])

plt.title("CatBoost: Top 10 Feature Importances")
plt.gca().invert_yaxis()  # Invert the y-axis to display the most important feature at the top
plt.show()
```
