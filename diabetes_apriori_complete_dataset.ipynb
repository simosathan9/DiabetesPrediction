```python
import pandas as pd
pd.set_option('display.max_columns', 500)
diabetes = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
diabetes
```

```python
diabetes.columns
```

```python
apriori_features = ['Diabetes_012', 'Smoker', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'Sex', 'Age', 'Education',
       'Income']
```

```python
# apriori and association rules
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

diabetes_apriori_df = diabetes[apriori_features]
diabetes_apriori_df['Diabetes_012'] = diabetes['Diabetes_012'].astype('category')
diabetes_apriori_df['Education'] = diabetes['Education'].astype('category')
diabetes_apriori_df['Income'] = diabetes['Income'].astype('category')
diabetes_apriori_df['Age'] = diabetes['Age'].astype('category')
diabetes_apriori_df = pd.get_dummies(diabetes_apriori_df)
diabetes_apriori_df = diabetes_apriori_df.astype('bool')
```

```python
# Apply the Apriori algorithm
frequent_itemsets = apriori(diabetes_apriori_df, min_support=0.1, use_colnames=True)

# Check if 'support' column exists
print(frequent_itemsets.head())

# Generate association rules (this should work if the support column is present)
rules = association_rules(frequent_itemsets, 10, metric="confidence", min_threshold=0.7)
```

```python
pd.set_option('display.max_colwidth', 100)
good_rules = rules[
    (rules['confidence'] > 0.9) &  # High confidence
    (rules['lift'] > 1.15) &        # Positive association
    (rules['support'] > 0.1) &     # Sufficient support
    (rules['conviction'] > 1) &    # Stronger association
    (rules['leverage'] > 0)        # Positive leverage indicates stronger relation
]

good_rules
```
